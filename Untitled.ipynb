```python
#creating bars 
```

```python
you can use the bar() function to draw the bar graphs
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x,y)
plt.show()
```

```python
x=["Apples","Mango"]
y=[400,600]
plt.bar(x,y)
```

```python
#horizontal Bars
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "red")
plt.show()
```

```python
#The bar() takes the keyword argument width to set the width of the bars 
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
x=np.array(["A","B","C","D"])
y=np.array([3,4,5,6])
plt.bar(x,y,width=0.1)
plt.show()
```

```python
#bar height
```

```python
import matplotlib.pyplot as plt 
import numpy as np
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.1)
plt.show()
```
